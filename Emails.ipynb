```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk.data
import nltk
nltk.download('punkt')
nltk.download('stopwords')
import re

stop = stopwords.words('english')

Data=df['Email_message'].head(10000)

Data['Message']=Data.apply(str)
#Data.Message = Data.Message.str.replace(' ', '')
print(Data['Message'])
Data['Stop'] = (Data['Message'].apply(lambda x: " ".join(x.lower() for x in str(x).split() \
                                   if x not in stop)))
Data['Stop']=Data['Stop'].apply(str)
```

```python
import string
Data['Stop'] = Data['Stop'].str.lower()#changing all to lowercase
print(Data['Stop'])
```

```python
Data['Stop'] = Data['Stop'].str.replace(r'https?://[^\s<>"]+|www\.[^\s<>"]+', "") # Removing all https/links
#removing "'" in the stem
Data['Stop'] = Data['Stop'].replace("'", "")
Data['Stop'] = Data['Stop'].str.replace(r"can't","cannot") # Changing cant to cannot
Data['Stop'] = Data['Stop'].str.replace(r"n't","not") # Changing n't to not
Data['Stop'] = Data['Stop'].str.replace(r" m ","am") # Changing m to am
Data['Stop'] = Data['Stop'].str.replace(r"I'm","I am") # Changing m to am
Data['Stop'] = Data['Stop'].str.replace(r"\'re","are") # Changing m to am
Data['Stop'] = Data['Stop'].str.replace(r"\'d","would") # Changing m to am
Data['Stop'] = Data['Stop'].str.replace(r"\'ll","will") # Changing m to am
print(Data['Stop'])





#Data['Stop'] = re.sub(r"can't","cannot",Data['Stop'])
#Data['Stop'] = re.sub(r"n't","not",Data['Stop'])
#Data['Stop'] = re.sub(r" m ","am",Data['Stop'])




```

```python
Data['Stop'].dropna(inplace=True) #Removing all null values
Data['Test'] = Data['Stop'].str.replace(r'[^\w\s]+', '')#replacing all punctuations with space
Data['tokens'] = Data['Test'].apply(word_tokenize)#tokenizing the top 10000 rows of the dataframe
print(Data['tokens'])


#from nltk.tokenize import RegexpTokenizer
#token = RegexpTokenizer(r'((?<=[^\w\s])\w(?=[^\w\s])|(\W))+', gaps=True)
#df['tokens']=token.tokenize(df['tokens_test'])
```

```python
#to do lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')

Data['lemmatize'] = Data['tokens'].head(10000).apply(
                    lambda lst:[lemmatizer.lemmatize(word) for word in lst])
print(Data['lemmatize'])
```

```python
print(Data.lemmatize[1])
```

```python
#to do stemming to get the root word
import nltk
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
Data['stem'] = Data['lemmatize'].head(10000).apply(
                    lambda lst:[porter_stemmer.stem(word) for word in lst])
print(Data['stem'])

```

```python
from gensim import corpora
import gensim
# storing the extracted tokens into the dictionary
my_dictionary = corpora.Dictionary(Data['stem'])
print(my_dictionary)
count = 0
for k, v in my_dictionary.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
```

```python
bow_corpus = [my_dictionary.doc2bow(doc) for doc in Data['stem']]
bow_corpus[4310]
```

```python
lda_model = gensim.models.LdaMulticore(bow_corpus, num_topics=10, id2word=my_dictionary, passes=2, workers=2)
print(lda_model)
```

```python
lda_model.show_topics(num_topics=10, num_words=10, log=False, formatted=True)
```

```python
from gensim.parsing.preprocessing import (preprocess_string, strip_punctuation,
strip_numeric)

lda_topics = lda_model.show_topics(num_words=10)

topics = []
filters = [lambda x: x.lower(), strip_punctuation, strip_numeric]

for topic in lda_topics:
    print(topic)
    topics.append(preprocess_string(topic[1], filters))

print(topics)
```
